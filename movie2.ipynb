```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['original_language'] = movies['original_language'].apply(lambda x: [lang.strip("'") for lang in x.split(',')])
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!= 3:
            L.append(i['name'])
            counter+=1
        else: break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
#stores names of first three cast members 
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(movies_new['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#to extract the root word from all other forms of words like love,loved,loving
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_new['tags']= movies_new['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#similarity gives vakues between 0 and 1 and if we import distances it will guve general values
```

```python
similarity=cosine_similarity(vectors)
similarity
#indicates similarity of each vector with every other vector 
#therefore for every vector there are 4036 options
```

```python
import joblib
joblib.dump(similarity, 'similarity_matrix.pkl')
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
#sort it in reverse order to obtain vector with maximum resemblemnce
```

```python
def recommend(movie):
    movie_index=movies_new[movies_new['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies_new.iloc[i[0]].title)
    
```

```python
recommend('Mom')
```
